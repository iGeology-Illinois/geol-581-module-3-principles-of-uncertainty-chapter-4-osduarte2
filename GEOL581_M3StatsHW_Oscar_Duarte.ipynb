A rock slope is to be cut as part of a road construction project. However, there is a risk that the slope could slide along the bedding planes into the excavation pit. In order to carry out a stability analysis it is necessary to estimate the angle of friction along the bedding planes. An initial assessment on the mean value of the friction angle can be obtained from the spectrum of possible manifestations of the friction angle. It cannot be smaller than the base friction angle, i.e. the friction angle that would result on the sawn, smooth surface of a specimen. The base friction angle for the greywackes that make up the slope is about 20°. The maximum possible friction angle can be estimated by back-calculating past failure cases. So far, no friction angle exceeding 35° has been back-calculated. What is the probability that the friction angle is smaller than 25°? Since no further information is available, the PDF of the friction angle can be assumed to be uniform.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt #Import matplotlib for plotting

# Defining the bounds and desired angle

lower_bound = 20
upper_bound = 35
desired_angle = 25

#Create a uniform distribution object
uniform_dist = stats.uniform(loc=lower_bound, scale=upper_bound - lower_bound)

# Calculate the probability that the friction angle is less than 25
probability = uniform_dist.cdf(desired_angle)
print(f"The probability of friction angle < {desired_angle} is: {probability}")

# Generate x values for the plot
x = np.linspace(lower_bound - 5, upper_bound + 5, 1000)

# Calculate the PDF values for the x values
pdf_values = uniform_dist.pdf(x)


# Plot the PDF
plt.plot(x, pdf_values, label='Uniform PDF')
plt.xlabel('Friction Angle (degrees)')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Uniform Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# New friction angle measurements (in degrees)
friction_angles_deg = np.array([25.4, 24.7, 21.6, 26.4, 25.1])

# Transform friction angles to radians (for completeness if further trig calculations are needed)
friction_angles_rad = np.deg2rad(friction_angles_deg)

# Apply natural logarithm transformation
log_transformed_angles = np.log(friction_angles_deg)

# Print results
print("Original friction angles (degrees):", friction_angles_deg)
print("Friction angles (radians):", friction_angles_rad)
print("Log-transformed friction angles:", log_transformed_angles)
```

```python
import numpy as np

# Prior information
mu_lnphi = 3.25
sigma_lnphi = 0.65
se_mu_lnphi = 0.15

# New sample data (log-transformed)
log_transformed_angles = np.array([3.2347, 3.2068, 3.0727, 3.2734, 3.2229])
n = len(log_transformed_angles)
# 1. Calculate the sample mean
sample_mean_lnphi = np.mean(log_transformed_angles)

# 2. Calculate the posterior mean
posterior_mean_lnphi = (
    (mu_lnphi / (se_mu_lnphi**2)) +
    (n * sample_mean_lnphi / (sigma_lnphi**2))
    ) / (
        (1 / (se_mu_lnphi**2)) +
        (n / (sigma_lnphi**2))
    )
print(f"Posterior Mean of Log-Transformed Friction Angle: {posterior_mean_lnphi:.4f}")
# 3. Calculate the posterior standard error of the mean
posterior_se_lnphi = np.sqrt(1 / ((1 / (se_mu_lnphi**2)) + (n / (sigma_lnphi**2))))
print(f"Posterior Standard Error of Log-Transformed Friction Angle: {posterior_se_lnphi:.4f}")
# 4. Estimate the mean friction angle
mean_phi = np.exp(posterior_mean_lnphi + (sigma_lnphi**2) / 2)
print(f"Estimated Mean Friction Angle: {mean_phi:.4f}")
# 5. Approximate standard error of the mean friction angle
se_mean_phi = mean_phi * posterior_se_lnphi
print(f"Approximate Standard Error of Mean Friction Angle: {se_mean_phi:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given shoot counts of sea thrift along a profile
shoot_counts = np.array([3, 5, 11, 12, 8, 19, 22, 18, 11, 13])

# Assuming measurements are taken in 20m x 20m squares, the distance between each measurement is a multiple of 20m.
# Define the lag distances
lag_distance = 20 # meters
num_counts = len(shoot_counts)
max_lag = num_counts - 1 # max lag of 9

# Calculate the distances for all pairs
distances = np.arange(1, num_counts) # distances of 20m, 40m, 60m, etc...

# Create an empty list to store semivariances
semivariances = []
lags = []

# Calculate the semivariance for each lag
for lag in distances: # lag can be 1, 2, 3, ...9.
    lag_values = []  # store the squared difference at a given lag.
    for i in range(num_counts - lag):  # compute only pairs within the max distance
        z_x = shoot_counts[i]  # value at location i
        z_x_h = shoot_counts[i + lag]  # value at location i + lag
        squared_difference = (z_x - z_x_h)**2 # compute the squared difference
        lag_values.append(squared_difference) # store the squared difference
    if lag_values:
         semivariance = 0.5 * np.mean(lag_values) # compute the semivariance at the given lag.
         semivariances.append(semivariance) # store the semivariance
         lags.append(lag*lag_distance) # store the distance in meters

# Plot the experimental semivariogram
plt.figure(figsize=(10, 6))
plt.plot(lags, semivariances, marker='o', linestyle='-')
plt.title("Experimental Semivariogram for Sea Thrift Shoot Counts")
plt.xlabel("Lag Distance (m)")
plt.ylabel("Semivariance")
plt.grid(True)
plt.show()

# Print the calculated semivariances for each lag distance.
for i, lag in enumerate(lags):
    print(f"Lag Distance: {lag:.0f} m, Semivariance: {semivariances[i]:.2f}")
```

```python
# Use this cell to load and process the orientation data.
# HINT:
# 1. Convert the dip direction/dip pairs into unit vectors in 3D space.
# 2. Sum these vectors, and derive the resultant direction and magnitude.
# 3. Calculate relevant statistics (standard deviation, confidence intervals) on a sphere.

import numpy as np

# Example structure:
# data = [(123,25), (147,22), (120,24), ...]
# Then convert each (alpha, beta) to x,y,z components on the unit sphere.
# Summation and geometry yield the mean vector direction and length.
#
# You might consider using spherical to Cartesian conversions:
# x = cos(dip) * sin(direction)
# y = cos(dip) * cos(direction)
# z = sin(dip)
# (in radians)
```

```python
import numpy as np

# Orientation data (dip direction/dip in degrees)
data = [(123, 25), (147, 22), (120, 24), (111, 24), (142, 26),
        (133, 27), (135, 22), (156, 21), (110, 25), (101, 25),
        (133, 20), (123, 22), (145, 25), (126, 24)]

# Convert dip direction and dip to radians
data_rad = [(np.deg2rad(dd), np.deg2rad(dip)) for dd, dip in data]

# Initialize sum of x, y, z components
sum_x = 0
sum_y = 0
sum_z = 0

# Convert to unit vectors and sum
for dip_dir_rad, dip_rad in data_rad:
  x = np.cos(dip_rad) * np.sin(dip_dir_rad)
  y = np.cos(dip_rad) * np.cos(dip_dir_rad)
  z = np.sin(dip_rad)
  sum_x += x
  sum_y += y
  sum_z += z

# Calculate the number of measurements
n = len(data)

# Calculate the mean vector components
mean_x = sum_x / n
mean_y = sum_y / n
mean_z = sum_z / n

# Calculate the magnitude of the resultant vector
resultant_magnitude = np.sqrt(sum_x**2 + sum_y**2 + sum_z**2)

# Calculate the mean resultant length
mean_resultant_length = resultant_magnitude / n

# Print the mean resultant length
print(f"Mean Resultant Length: {mean_resultant_length:.4f}")
```

```python
import numpy as np
from scipy.stats import chi2

# Orientation data (dip direction/dip in degrees)
data = [(123, 25), (147, 22), (120, 24), (111, 24), (142, 26),
        (133, 27), (135, 22), (156, 21), (110, 25), (101, 25),
        (133, 20), (123, 22), (145, 25), (126, 24)]

# Convert dip direction and dip to radians
data_rad = [(np.deg2rad(dd), np.deg2rad(dip)) for dd, dip in data]

# Initialize sum of x, y, z components
sum_x = 0
sum_y = 0
sum_z = 0

# Convert to unit vectors and sum
for dip_dir_rad, dip_rad in data_rad:
  x = np.cos(dip_rad) * np.sin(dip_dir_rad)
  y = np.cos(dip_rad) * np.cos(dip_dir_rad)
  z = np.sin(dip_rad)
  sum_x += x
  sum_y += y
  sum_z += z

# Calculate the number of measurements
n = len(data)

# Calculate the mean vector components
mean_x = sum_x / n
mean_y = sum_y / n
mean_z = sum_z / n

# Calculate the magnitude of the resultant vector
resultant_magnitude = np.sqrt(sum_x**2 + sum_y**2 + sum_z**2)

# Calculate the mean resultant length
mean_resultant_length = resultant_magnitude / n

# Calculate dip direction
dip_direction_rad = np.arctan2(mean_x, mean_y)

# Correct dip direction to be between 0 and 2*pi (or 0 and 360 degrees)
if dip_direction_rad < 0:
    dip_direction_rad = dip_direction_rad + 2*np.pi

# Convert dip direction to degrees
dip_direction_deg = np.rad2deg(dip_direction_rad)

# Calculate dip
dip_rad = np.arcsin(mean_z)

# Convert dip to degrees
dip_deg = np.rad2deg(dip_rad)

# --- NEW CODE FOR SPHERICAL STANDARD DEVIATION AND CONFIDENCE INTERVAL ---

# Calculate the spherical standard deviation (kappa)
if mean_resultant_length < 1:
    kappa = (n - 1) / (n * (1- mean_resultant_length))
else:
    kappa = 1e9 # Arbitrary large number, if mean resultant length is 1.

# Calculate the angular standard deviation (in degrees)
angular_std_dev_rad = np.sqrt(2*(1-mean_resultant_length))
angular_std_dev_deg = np.rad2deg(angular_std_dev_rad)

# Calculate the critical value of the chi-squared distribution
alpha = 0.1
critical_chi2 = chi2.ppf(1 - alpha, 2)

# Calculate the radius of the confidence cone (in radians)
if mean_resultant_length < 1:
    confidence_radius_rad = np.sqrt((critical_chi2/(n*kappa)))
else:
    confidence_radius_rad = 0 #if mean resultant length is 1.

# Convert confidence radius to degrees
confidence_radius_deg = np.rad2deg(confidence_radius_rad)


# Print the results
print(f"Mean Resultant Length: {mean_resultant_length:.4f}")
print(f"Dip Direction of the Mean Vector: {dip_direction_deg:.2f} degrees")
print(f"Dip of the Mean Vector: {dip_deg:.2f} degrees")
print(f"Spherical Standard Deviation: {angular_std_dev_deg:.2f} degrees")
print(f"Spherical Confidence Interval (Radius): {confidence_radius_deg:.2f} degrees (alpha = {alpha})")
```
